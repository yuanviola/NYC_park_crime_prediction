```python
import pandas as pd
import geopandas as gpd
```

```python
park_crime=pd.read_csv('../data/1_parkcrime.csv')
```

```python
shape_path="../data/2 park_location.geojson"
park_shape = gpd.GeoDataFrame.from_file(shape_path).to_crs(epsg=4326)
```

```python
park_shape.park_name= park_shape.park_name.str.lower()
```

```python
location = park_shape[['park_name','geometry']]
location = location.rename(columns={'park_name':'PARK'})
```

```python
score = pd.read_csv('../data/3_park_score.csv')
score = score[['SiteName','average_score']]
score = score.rename(columns={"SiteName":'PARK'})
```

```python
score.PARK=score.PARK.str.lower()
```

```python
len(park_crime.PARK.value_counts())
```

#### We have crime record for 743 parks in total.

```python
len(park_crime.merge(score,on='PARK',how='inner')['PARK'].value_counts())
```

#### If we use park score, inner merge, only 492 parks left.

```python
park_crime.merge(score,on='PARK',how='outer')
```

```python
demo = pd.read_csv('../data/4 cleaned_medianincome_houldhold.csv')
demo = demo[['park_name','household','median_income']]
demo= demo.rename(columns={'park_name':'PARK'})
demo.PARK = demo.PARK.str.lower()
```

```python
park_crime = park_crime.merge(demo,on='PARK',how = 'inner')
```

```python
complaints = pd.read_csv('../data/6 311_complains_merged_with_crime.csv')
del complaints['Unnamed: 0']
```

```python
crime_311_income = complaints.merge(demo,on='PARK',how='outer')
```

```python
crime_311_income_score = crime_311_income.merge(score,on='PARK')
crime_311_income_score.head()
```

```python
public_h = pd.read_csv('../data/7 cleaned nyc public house.csv')
del public_h['Unnamed: 0']
```

```python
public_h= public_h.rename(columns={'park_name':"PARK"})
```

```python
public_h.PARK = public_h.PARK.str.lower()
```

```python
crime_311_income_score_pubh =crime_311_income_score.merge(public_h,on='PARK')
```

```python
crime_311_income_score_pubh = crime_311_income_score_pubh[crime_311_income_score_pubh.CATEGORY=='ONE ACRE OR LARGER']
```

```python
crime_311_income_score_pubh = crime_311_income_score_pubh.fillna(0)
```

```python
crime_311_income_score_pubh = crime_311_income_score_pubh.drop_duplicates() 
```

```python
crime_311_income_score_pubh= crime_311_income_score_pubh.reset_index(drop= True)
```
